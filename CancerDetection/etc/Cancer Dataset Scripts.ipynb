## Trial Plot of Reduced

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.colors as clr
import numpy

# remove weights since we don't know what to do with them yet
reduced_noWeights = reduced.iloc[1:,4:]
dim = reduced_noWeights.shape

# ranges (for verification)
rowend = dim[0]
colend = dim[1]

### !! Use this smaller set to compare against excel document !! ###
## Adjust plt.scatter option vmin = -1, vmax = 1 for clarity ##
#rowend = 10
#colend = 5

# prepare meshgrid and color data
rows = list(range(0,rowend))
cols = list(range(0,colend))
x, y = numpy.meshgrid(rows,cols)
color = reduced_noWeights.iloc[rows,cols].values # maintain log2 transform

# plot
cdict = clr.LinearSegmentedColormap.from_list('RdGn', [(0,1,0),(0,0,0),(1,0,0)], N=256, gamma=1.0)

plt.figure(figsize=(8,8))
plt.scatter(y+1, x+1, c=color[x,y], cmap=cdict, vmin=-2, vmax=2, lw=0, s=10)

# zoom in
plt.gca().set_xlim([0,colend+1])
plt.gca().set_ylim([0,rowend+1])

# format as microarray
plt.gca().xaxis.tick_top()
plt.gca().xaxis.set_label_position('top') 
plt.gca().invert_yaxis()
plt.gca().set_axis_bgcolor((1,1,1))

# add information
plt.colorbar()
plt.xlabel('Sample')
plt.ylabel('Gene')

plt.show()

# commentary
print('cancer expressed (red), normal expressed (green)')
print('scale from 1/4 to 4-fold expression ratio (cancer/normal)')
```
